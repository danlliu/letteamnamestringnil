```python
import requests
import urllib.request
from bs4 import BeautifulSoup
import json
```

```python
spells = {
    "cantrip": [],
    "1st level": [],
    "2nd level": [],
    "3rd level": [],
    "4th level": [],
    "5th level": [],
    "6th level": [],
    "7th level": [],
    "8th level": [],
    "9th level": []
}
legend = {
    "V": "verbal component (spoken incantation)",
    "S": "somatic component (hand motion)",
    "M": "material (uses up a certain material to cast)"
}
class_spells = {
    "Artificer": {},
    "Bard": {},
    "Cleric": {},
    "Druid": {},
    "Eldritch Knight": {},
    "Paladin": {},
    "Ranger": {},
    "Sorcerer": {},
    "Warlock": {},
    "Wizard": {}
}
# print(class_ids)
```

```python
for idx, cls in enumerate(class_spells.keys()):
    pages = [
        "http://dnd5e.wikidot.com/spells:artificer",
        "http://dnd5e.wikidot.com/spells:bard",
        "http://dnd5e.wikidot.com/spells:cleric",
        "http://dnd5e.wikidot.com/spells:druid",
        "http://dnd5e.wikidot.com/spells:eldritch-knight",
        "http://dnd5e.wikidot.com/spells:paladin",
        "http://dnd5e.wikidot.com/spells:ranger",
        "http://dnd5e.wikidot.com/spells:sorcerer",
        "http://dnd5e.wikidot.com/spells:warlock",
        "http://dnd5e.wikidot.com/spells:wizard"
    ]
    page = urllib.request.urlopen(pages[idx])
    soup = BeautifulSoup(page.read().decode(), "html.parser")
    spell_class_dict = {}
    min_lvl = 0
    max_lvl = 10
    if cls == "Artificer":
        max_lvl = 6
    if cls == "Eldritch Knight":
        max_lvl = 5
    if cls == "Paladin":
        min_lvl = 1
        max_lvl = 6
    if cls == "Ranger":
        min_lvl = 1
        max_lvl = 6
    
    for i in range (min_lvl, max_lvl):
        spell_subdict = {}
        div_id = "wiki-tab-0-" + str(i)
        spell_div = soup.findAll("div", {"id": div_id})
        for spells in spell_div:
            for spell in spells.findChildren("tr")[1:]:
                lnk = str(spell.find("a")["href"])
                sp = spell.get_text().split("\n")
                sp_dict = {
                    "Spell Name": sp[1],
                    "Spell Link": "http://dnd5e.wikidot.com" + lnk,
                    "School": sp[2],
                    "Cast Time": sp[3],
                    "Range": sp[4],
                    "Duration": sp[5],
                    "Components": sp[6]
                }
                sp[3] = sp[3].title()
                if not (sp[3].endswith("tion") or sp[3].endswith("Minute") or sp[3].endswith("Minutes")
                    or sp[3].endswith("Hour") or sp[3].endswith("Hours") or sp[3].endswith("Days") or sp[3] == "Special" or sp[3] == "Self"):
                        split_str = sp[3].rsplit(" ", 1)
                        sp_dict["Cast Time"] = split_str[0]
                        sp_dict["Special Type"] = split_str[1].upper()
                spell_subdict[sp[1]] = sp_dict
                #for idx, line in enumerate(spell):
                    #print(idx, line.string)
        if i == 0:
            spell_class_dict["Cantrip"] = spell_subdict
        else:
            tag = "Level " + str(i)
            spell_class_dict[tag] = spell_subdict
    class_spells[cls] = spell_class_dict
    
print(spell_class_dict)
```

```python
def clsSpells(cls, allowHB):
    url = "http://dnd5e.wikidot.com/spells:" + str(cls)
    test_page = urllib.request.urlopen(url)
    cont = test_page.read().decode()
    #print(cont)
    test_soup = BeautifulSoup(cont, "html.parser")
    #print(test_soup.prettify())

    spell_dict = {}
    for i in range (0, 10):
        spell_subdict = {}
        div_id = "wiki-tab-0-" + str(i)
        spell_div = test_soup.findAll("div", {"id": div_id})
        for spells in spell_div:
            for spell in spells.findChildren("tr")[1:]:
                lnk = str(spell.find("a")["href"])
                sp = spell.get_text().split("\n")
                sp_dict = {
                    "Spell Name": sp[1],
                    "Spell Link": "http://dnd5e.wikidot.com" + lnk,
                    "School": sp[2],
                    "Cast Time": sp[3],
                    "Range": sp[4],
                    "Duration": sp[5],
                    "Components": sp[6]
                }
                
                if not allowHB:
                    if "(HB)" in sp_dict["Spell Name"]:
                        continue
                
                sp[3] = sp[3].title()
                if not (sp[3].endswith("tion") or sp[3].endswith("Minute") or sp[3].endswith("Minutes")
                    or sp[3].endswith("Hour") or sp[3].endswith("Hours") or sp[3].endswith("Days") or sp[3] == "Special" or sp[3] == "Self"):
                        #print(str(i) + " " + sp[1] + " " + sp[3])
                        split_str = sp[3].rsplit(" ", 1)
                        sp_dict["Cast Time"] = split_str[0]
                        sp_dict["Special Type"] = split_str[1].upper()
                spell_subdict[sp_dict["Spell Name"]] = sp_dict
                #for idx, line in enumerate(spell):
                    #print(idx, line.string)
        if i == 0:
            spell_dict["Cantrip"] = spell_subdict
        else:
            tag = "Level " + str(i)
            spell_dict[tag] = spell_subdict
            
    return spell_dict
```

```python
classes = [
        "artificer",
        "bard",
        "cleric",
        "druid",
        "eldritch-knight",
        "paladin",
        "ranger",
        "sorcerer",
        "warlock",
        "wizard"
    ]

allowHB = False 
all_spells = {}
for cls in classes:
    all_spells[cls] = clsSpells(cls, allowHB)
    
with open("scraper_output/wikidot_all_spells.json", "w", encoding='utf-8') as outfile:
    json.dump(all_spells, outfile, ensure_ascii=False, indent=4)
```
